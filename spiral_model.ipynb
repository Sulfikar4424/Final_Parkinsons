```python
from flask import Flask,request, url_for, redirect, render_template
import pickle
import numpy as np
from tensorflow import keras
import cv2
import matplotlib.pyplot as plt
```

```python
model=keras.models.load_model('parkinson_disease_detection.h5')
```

```python
# Testing Model on Images
labels = ['Healthy', 'Parkinson']
image = cv2.imread('test_image_parkinson.png')
```

```python
image = cv2.resize(image, (128,128))
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image = np.array(image)
image = np.expand_dims(image, axis=0)
image = np.expand_dims(image, axis=-1)
```

```python
ypred_healthy = model.predict(image)
```

```python
figure = plt.figure(figsize=(2, 2))
image = np.squeeze(image, axis=0)
image = np.squeeze(image, axis=-1)
plt.imshow(image)
plt.axis('off')
plt.title(f'Prediction by the model: {labels[np.argmax(ypred_healthy[0], axis=0)]}')
plt.show()
```
